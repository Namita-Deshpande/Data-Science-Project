```python
#removing outlier (noise in data)
import scipy.stats as stats
import numpy as np
z_scores = np.abs(stats.zscore(x))
threshold =3
print("size before removing outliers:",x.shape)
outliers_df = x[(z_scores > threshold).any(axis=1)]
x=x[(z_scores <= threshold).all(axis=1)]
print("size before removing outliers:",x.shape)
```

```python
y=x['Survived'] #extracting last column as target features(dependent feature)
y   
```

```python
z=x.drop('Survived',axis=1) #extracting all column expect target column as independent feauture

z
```

```python
# Split our data between a training set and a test set:
z_train, z_test, y_train, y_test = train_test_split(z, y,test_size=0.30,
                                                    stratify=y, random_state=66)
```

```python
correlation=x.corr()['Survived']
correlation
```

```python
sorted_features=correlation.abs().sort_values(ascending=False)
sorted_features.head(10)
```

```python
#Random forest classifier
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=100,random_state=100)

#fit model
forest.fit(z_train,y_train)

#prediction 
pred = forest.predict(z_test)

#predicted labels
pred[:10]

#Accuracy percentage
accuracy_percentage=100*np.sum(pred==y_test)/len(y_test)
accuracy_percentage
```

```python
auc=sklearn.metrics.roc_auc_score(y_test,pred)
print("auc",auc)
#AUC
```
